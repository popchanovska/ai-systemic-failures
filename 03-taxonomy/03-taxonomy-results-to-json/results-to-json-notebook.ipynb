```python
import json

# Input JSONL file (result jsonl from Batch API)
jsonl_file = "INPUT.jsonl"

# Output JSON file
json_file = "OUTPUT.json"

all_taxonomies = []

# Read JSONL file line by line
with open(jsonl_file, "r", encoding="utf-8") as f:
    for line in f:
        data = json.loads(line)
        content_str = data.get("response", {}).get("body", {}).get("choices", [])[0].get("message", {}).get("content")
        if content_str:
            try:
                content_json = json.loads(content_str)
                derived_taxonomy = content_json.get("derived_taxonomy")
                if derived_taxonomy:
                    all_taxonomies.append(derived_taxonomy)
            except json.JSONDecodeError:
                print("Skipping a line due to error")

# Save results to a single JSON file
with open(json_file, "w", encoding="utf-8") as f:
    json.dump(all_taxonomies, f, indent=2, ensure_ascii=False)

print(f"Saved {len(all_taxonomies)} derived_taxonomy entries to {json_file}")
```
